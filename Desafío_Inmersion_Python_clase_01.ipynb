**Paso 01**

Cargar y Explorar los Datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import warnings

drive.mount("/content/Drive")
warnings.filterwarnings("ignore")
```

```python
# Montar Google Drive
drive.mount('/content/drive', force_remount=True)

# Especificar la ruta del archivo en Google Drive
file_path = '/content/drive/MyDrive/Inmersión Python/german_credit.csv'

# Cargar el archivo CSV
df = pd.read_csv(file_path)
```

**Paso 02**

Para realizar el paso dos se puede utilizar gráficos de cajas (box plots) para detectar valores atípicos.

```python
# Verificar si el DataFrame df está definido
if 'df' in globals():
    # Graficar los box plots de las columnas numéricas
    plt.figure(figsize=(15, 10))
    df.boxplot()
    plt.xticks(rotation=90)
    plt.title("Box Plots de las Columnas Numéricas")
    plt.show()
else:
    print("El DataFrame df no está definido. Asegúrate de que el archivo CSV se haya cargado correctamente.")
```

**Paso 03**

Crear un Mapa de Calor para Correlación


```python
# Verificar si el DataFrame df está definido
if 'df' in globals():
    # Crear el mapa de calor de correlación
    plt.figure(figsize=(15, 10))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Mapa de Calor de Correlación")
    plt.show()
else:
    print("El DataFrame df no está definido. Asegúrate de que el archivo CSV se haya cargado correctamente.")
```

**Paso 04**

Generar Histogramas y Extraer Conclusiones

```python
# Verificar si el DataFrame df está definido
if 'df' in globals():
    # Graficar histogramas de todas las columnas numéricas
    df.hist(figsize=(20, 20), bins=20, edgecolor='black')
    plt.suptitle("Histogramas de las Columnas Numéricas", fontsize=20)
    plt.show()
else:
    print("El DataFrame df no está definido. Asegúrate de que el archivo CSV se haya cargado correctamente.")
```
